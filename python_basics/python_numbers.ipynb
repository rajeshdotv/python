In Python, there are three distinct numeric types:

1. Integers (Eg: 10)
2. Floating-point numbers (Eg: 1.5)
3. Complex numbers (Eg: 3+4j)

## Integers
An integer is a whole number that can be positive or negative.

```python
x = 10
y = -10
z = 123456789
```

```python
print(f"{type(x)}, {type(y)}, {type(z)}")
```

In Python 3, there is no limit to how long an integer value can be. It can grow to have as many digits as your computer’s memory space allows.

```python
# Integers have unlimited precision
x = 99999999999999999999999999999999999999999999999999999999999999999999999999999999
```

We normally write integers in base 10. <br> However, Python allows us to write integers in Hexadecimal (base 16), Octal (base 8), and Binary (base 2) formats. <br>You can do that by adding one of the following prefixes to the integer.

```python
# Base - 2, Interpretation - Binary, Prefix - ‘0b’ or ‘0B’
# Base - 8, Interpretation - Octal, Prefix - ‘0o’ or ‘0O’
# Base - 16, Interpretation - Hexadecimal, Prefix - ‘0x’ or ‘0X’
```

```python
# Integers in binary, octal, hexadecimal formats

# binary
print(0b10111011) # Prints 187

# octal
print(0o10) # prints 8

# hex
print(0xFF) # prints 255
```

## Floating-Point Numbers
Floating-point number or Float is a positive or negative number with a fractional part.

```python
x = 10.1
y = -20.4
z = 1.23456789
```

```python
print(f"{type(z)}, {type(y)}, {type(z)}")
```

You can append the character e or E followed by a positive or negative integer to specify scientific notation.

```python
# Scientific Notation
print(32e3) # prints 32000.0

print(4.2e-3) # prints 0.0042
# 4.2 * 10^-3
```

The maximum value a float can have is approximately 1.8×10^308 . Any number greater than that is indicated by the string inf (infinity)

```python
# Maximum value of float
print(1.79e308) # prints 1.79e+308

print(1.8e308) # prints inf
```

the minimum value a float can have is approximately 5.0×10^-324 . Any number, less than that is considered zero.

```python
# Minimum value of a float
print(5e-324) # prints 5e-324

print(5e-325)
```

## Complex Numbers
A complex number is specified as real_part + imaginary_part, where the imaginary_part is written with a j or J.

```python
x = 2j
y = 3+4j
```

```python
# To extract real and imaginary parts from a complex number x, use x.real and x.imag
x = 3+4j

# real part
print(x.real)

# imaginary part
print(x.imag)
```
