## 4. Counting and printing Vowels in a Given Word

```python
name = "rajeshkumar"
vowels = ['a', 'e', 'i', 'o', 'u']
vowels_count = 0
vowels_list = []
```

```python
for i in name:
  if i in vowels:
    vowels_list.append(i)
    vowels_count += 1
print(vowels_count)
print(vowels_list)
```

## 5. Counting and printing Consonants in a Given Word

```python
name = "rajeshkumar"
vowels = ['a', 'e', 'i', 'o', 'u']
consonants_count = 0
consonants_list = []
```

## 7. Writing Fibonacci Series

0,1,1,2,3,5,8,12....



```python
def fibonaaci_series(n):
  n1 = 0
  n2 = 1
  count = 0
  if n <= 0:
    print("Please Enter positive number, as Classical Fibonaaci series is only available for positive number only")
  elif n == 1:
    print(n1)
  else:
    while count < n:
      print(n1)
      nth = n1 + n2
      n1 = n2
      n2 = nth
      count+=1
```

```python
fibonaaci_series(8)
```

```python
def fib_series_list(n):
  fib = [0, 1]
  for i in range(n):
    # Adding last term and previous term from list
    fib.append(fib[-1] + fib[-2])
    # Converting list of Integers to string
    fib_series = ', '.join([str(i) for i in fib])
    return fib_series
```

```python
fib_series_list(5)
```

```python
# Using Recursive
def fib_recursive(n):
  if n <= 1:
    return n
  else:
    return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
n = 8
for i in range(n):
  print(fib_recursive(i))
```

## 16. Counting Digits, Letters, and Spaces in a String

```python
# Without using Regex
def count_characters(input):
  digit_count = 0
  alphabet_count = 0
  special_char_count = 0
  for char in input:
    if char.isdigit():
      digit_count += 1
    elif char.isalpha():
      alphabet_count += 1
    else:
      special_char_count += 1
  return f"Digit count is {digit_count}, Alphabet count is {alphabet_count}, Special Characters count is {special_char_count}"
```

```python
input = 'Python is 1'
count_characters(input)
```

```python
# Using Regex
import re
input = 'Python is 1'
```

```python
digit_count = re.sub("[^0-9]", "", input)
letterCount = re.sub("[^a-zA-Z]", "", input)
spaceCount = re.findall("[ \n]", input)
print(len(digit_count))
print(len(letterCount))
print(len(spaceCount))
```

## 17. Counting Special Characters in a String

```python
sp_char = "!@#$%^&*()rajesh"
```

```python
def find_special_char_count(input):
  special_char_count = 0
  special_char_list = []
  for char in input:
    if not char.isdigit() and not char.isalpha():
      special_char_list.append(char)
      special_char_count += 1
  return special_char_list, special_char_count
```

```python
find_special_char_count(sp_char)
```
