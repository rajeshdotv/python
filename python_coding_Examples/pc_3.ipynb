```python
### Complete Code
input_str = "rajeshvelpuri"

vowels = ['a', 'e', 'i', 'o', 'u']
vowel_indices = [i for i, char in enumerate(input_str) if char.lower() in vowels]
reversed_vowels = ''.join([input_str[i] for i in reversed(vowel_indices)])
reversed_vowels_iter = iter(reversed_vowels)
print(reversed_vowels_iter)

output = ''.join([next(reversed_vowels_iter) if char.lower() in vowels else char for char in input_str])
print(output)
```

### 2. Create a Pandas Data Frame and Explain "loc" and "iloc"

loc is used for label-based indexing, where we specify row and column labels <br> while iloc is used for integer position-based indexing, where we specify row and column positions.

```python
import pandas as pd
```

```python
# Creating a sample Data frame
data = {
    'Name': ["Raj", "kumar", "grace", "keziah", "ramesh"],
    'Age': [31, 28, 26, 25, 21],
    'Gender': ['M', 'M', 'F', 'F', 'M'],
    'Location': ['Hyderabad', 'Secunderabad', 'Guntur', 'Guntur', 'Vijayawada']
}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
# loc is primarily label-based indexing. It is used to access a group of rows and columns by label(s) or a boolean array.
# Syntax: df.loc[row_label(s), column_label(s)]
```

```python
# Using 'loc' to select rows and columns by labels
print(df.loc[2]) # Selcts rows with label 2
```

```python
print(df.loc[1:3, ['Name', 'Age']]) # Prints rows with labels 1 to 3 and specified column names
```

```python
## iloc is primarily integer position-based indexing. It is used to access a group of rows and columns by integer position(s) or a boolean array.
## Syntax: df.iloc[row_index(s), column_index(s)]
```

```python
# Using 'iloc' to select rows and columns by Integer positon
print(df.iloc[2])
```

```python
print(df.iloc[1:3], [0, 1]) # Selects rows from 1 to 3 with columns which are in postion of 0 and 1
```

### 3. function to count the occurrences of each character in a string

```python
def count_characters(input_string):
  # Initialize dictionary where we keep count of characters in this dict
  char_count = {}
  for char in input_string:
    if char in char_count:
      char_count[char] += 1 # If char is already present in dict , increment its value
    else:
      char_count[char] = 1 # It char is not in dict, Initilize and keep the count as 1
  return char_count
```

```python
count_characters("RRajesh")
```

### 4. Write a function to remove duplicates from a list.

```python
def remove_duplicates_list(input_list):
  unique_list = []
  for i in input_list:
    if i not in unique_list:
      unique_list.append(i)
  return unique_list
```

```python
remove_duplicates_list([1, 1, 2, 3, 4, 5])
```

### 5.  function to find the maximum value in a dictionary:

```python
def find_max_value(dictionary):
  # Check if dict is empty
  if not dictionary:
    return None
  # Initialize the max value to the smallest Integer possible
  max_value = float('-inf')
  # By initializing max_value to negative infinity, we ensure that any value we encounter in the dictionary will be greater than max_value initially.
  # This allows us to correctly update max_value as we iterate through the dictionary's values.

  # Iterate through the values in Dictionary
  for value in dictionary.values():
    if value > max_value:
      max_value = value
  return max_value
```

```python
test_dict = {'a':5, 'b': 42, 'c': 37}
find_max_value(test_dict)
```

## 6. Write a function to merge two dictionaries.

```python
# Merge dictionaries using update method
def merge_dict(dict1, dict2):
  merged_dict = {}
  # Using update method
  merged_dict.update(dict1)
  merged_dict.update(dict2)
  return merged_dict
```

```python
# Merge dictionaries using dictionary comprehension
def merge_dict_comp(dict1, dict2):
  merge_dict = {**dict1, **dict2}
  return merge_dict
```

```python
d1 = {'a':1,'b':2}
d2 = {'c':3, 'd': 4}
merge_dict(d1, d2)
```

```python
merge_dict_comp(d1, d2)
```
