## Question 2

Question: Write a program which can compute the factorial of a given numbers. The results should be printed in a comma-separated sequence on a single line. Suppose the following input is supplied to the program: 8 Then, the output should be: 40320

Hints: In case of input data being supplied to the question, it should be assumed to be a console inpu

```python
# Using Recursive
def factorial_recursive(n):
  if n < 0:
    print("Enter Positive Number")
  elif n == 0:
    return 1
  else:
    return n * factorial_recursive(n-1)
```

```python
n = int(input("Enter the number"))
print(factorial_recursive(n))
```

```python
# Using Iteration
def factorial_iterative(n):
  factorial = 1
  for i in range(1, n+1):
    factorial = factorial * i # can be written as: factorial *= i
  return factorial
```

```python
factorial_iterative(8)
```

```python
# Using Math Library
import math
def factorial_math(n):
  factorial = math.factorial(n)
  return factorial
```

```python
factorial_math(8)
```

```python
def count_char_num(input):
  char_dict = {"DIGITS":0, "LETTERS":0}
  for c in input:
    if c.isalpha():
      char_dict['LETTERS'] += 1
    elif c.isdigit():
      char_dict['DIGITS'] += 1
    else:
      pass
  return char_dict
```

Question: Write a program that computes the net amount of a bank account based a transaction log from console input. The transaction log format is shown as following: D 100 W 200

D means deposit while W means withdrawal. Suppose the following input is supplied to the program: D 300 D 300 W 200 D 100 Then, the output should be: 500

Hints: In case of input data being supplied to the question, it should be assumed to be a console input.

```python
s = input()
```

```python
s
```

```python
values = s.split()
values
```

```python
def compute_net_amount(transaction_log):
  net_amount = 0
  transactions = transaction_log.split()

  for i in range(0, len(transactions), 2):
    transaction_type = transactions[i]
    transaction_amount = int(transactions[i+1])

    if transaction_type == 'D':
      net_amount += transaction_amount
    elif transaction_type == 'W':
      net_amount -= transaction_amount

  return net_amount
```

```python
transaction_log = input()
```

```python
compute_net_amount(transaction_log)
```

```python
def get_max(s1, s2):
  if len(s1) > len(s2):
    print(s1)
  elif len(s2) > len(s1):
    print(s2)
  elif len(s1) == len(s2):
    print(s1)
    print(s2)
  else:
    pass
```

```python
def binary_search(arr, target):
  arr.sort() # It will sort the list
  left = 0
  right = len(arr)-1
  while left <= right:
    #Calculate mid
    mid = (left + right)//2

    # Check if target is mid value
    if arr[mid] == target:
      return mid
    # check if mid < target
    elif arr[mid] < target:
      left = mid + 1 # Ignore first left half
    else:
      right = mid - 1
  return -1
```

```python
arr = [2, 5, 7, 9, 11, 17, 222]
target = 222
```

```python
binary_search(arr, target)
```

## Question 98

Please write a program which accepts a string from console and print the characters that have even indexes.

Example: If the following string is given as input to the program:

H1e2l3l4o5w6o7r8l9d

Then, the output of the program should be:

Helloworld

```python
s = 'H1e2l3l4o5w6o7r8l9d'
```

```python
''.join([s[i] for i,v in enumerate([c for c in s]) if i%2 == 0])
```

## Question 100

Please write a program - to reverse only vowels from a string.
consonants place shouldn't change

```python
name = "rajeshvelpuri"
```

```python
name_list = list(name)
```

```python
vowels = ['a','e','i','o','u']
```

```python
info = [(i,v) for i,v in enumerate(name) if v in vowels]
```

```python
info
```

```python
vowels_indices = [i[0] for i in info]
vowels_indices
```

```python
vowels_list = [i[1] for i in info]
vowels_list
```

```python
vowels_list_reverse = vowels_list[::-1]
vowels_list_reverse
```

```python
for i, index in enumerate(vowels_indices):
  name_list[index] = vowels_list_reverse[i]
```

```python
name_list
```

```python
''.join(name_list)
```
