```python
def find_second_lowest_students(records):
  # Sort the records based on score
  sorted_records = sorted(records, key = lambda x: x[1])
  # Find second Lowest Marks
  second_lowest_marks = sorted_records[1][1]
  # Find all Students with second lowest marks
  second_lowest_students = sorted([sublist[0] for sublist in sorted_records if sublist[1] == second_lowest_marks])
  return second_lowest_students
```

```python
if __name__ == '__main__':
  n = int(input())
  records = []
  if n >= 2 and n <= 5:
    for _ in range(n):
      temp_list = []
      name = input()
      score = float(input())
      temp_list.append(name)
      temp_list.append(score)
      records.append(temp_list)

  second_lowest_students = find_second_lowest_students(records)
  print('\n'.join(second_lowest_students))
```

## HR Question - 4

string = "ABCDCDCCDC"

sub_string = "CDC"

Count substring in string (From left to right only)

```python
def count_substring(string, substring):
  count = 0
  sub_len = len(substring)
  for i in range(len(string) - sub_len + 1):
    if string[i:i + sub_len] == substring:
      count += 1
  return count
```

```python
string = "ABCDCDC"
sub_string = "CDC"

count_substring(string, sub_string)
```

```python
def get_result(arr):
  count_word = {}
  for word in arr:
    if word in count_word:
      count_word[word] += 1
    else:
      count_word[word] = 1

  unique_arr = [item for index, item in enumerate(arr) if item not in arr[:index]]
  unique_word_count = len(unique_arr)

  count_list = []
  for i in unique_arr:
    count_list.append(str(count_word[i]))

  return unique_word_count, count_list
```
