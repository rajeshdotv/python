### 4.Write a program to find if the given number is prime or not.

```python
# a whole number greater than 1 that cannot be exactly divided by any whole number other than itself and 1 (e.g. 2, 3, 5, 7, 11).
#"prime numbers are very useful in cryptography"
```

```python
num = int(input("Enter the number: "))
flag = False
if num > 1:
  for i in range(2, num):
    if (num % i) == 0:
      flag = True
      break
if flag:
  print(f"The Entered number {num} is not a Prime Number")
else:
  print(f"The Entered number {num} is a Prime Number")
```

Addition (+)
Subtraction (-)
Multiplication (*)
Division (/)
Modulus (%) --> Gives Remainder
Exponentiation (**)
Floor Division (//)  --> Gives Quotient --> Round of to Nearest Integer. For Eg: 3.75, floor division gives value of 4

```python
num = int(input("Enter the number: "))
temp = num
reverse = 0
while temp > 0:
  remainder = temp % 10 # Gives Last Number for eg num: 12321, gives 1
  reverse = (reverse * 10) + remainder # Adds the number from last, For eg num: 12321 --> (0*10 + 1) = 1
  temp = temp // 10 # Removes the last number and give, for eg: 12321, gives 1232

# Constructing the reverse number using modulus,floor division
if num == reverse:
  print(f"The Entered number {num} is palindrome")
else:
  print(f"The Entered number {num} is not Palindrome")
```

Armstrong number is the number that equals the sum of its digits, each raised to a power. For example, if you have a number like 153, it's an Armstrong number because 1^3 + 5^3 + 3^3 equals 153

```python
num = 153
```

The mod_size function you provided returns the length of the string representation of an integer i. However, instead of using the str function to convert i into a string, it uses the string modulo operator %. While this approach works, it's a bit unconventional.

"%i" % i is a string formatting operation using the % operator. %i is a placeholder for an integer. It replaces %i with the value of i in the resulting string.

```python
def num_size(i):
    return len("%i" % i) # Uses string modulo instead of str(i)
```

```python
num_size(153)
```

```python
len("%i" % 153)
```

```python
print(range(3))
```

```python
num = int(input("Enter the Number: "))
sum = 0
for i in str(num):
  sum += int(i) ** num_size(num) # ** gives the power of the value
if num == sum:
  print(f"The Entered Number {num} is Armstrong number")
else:
  print(f"The Entered Number {num} is not Armstrong number ")
```

### 7. Different ways of finding length of a Number

```python
# Method 1 --> Converting Integer to string
def size_str(i):
  return len(str(i))
size_str(153)
```

```python
# Method 2 --> Base10 logarithm
import math
# In mathematics, the logarithm is the inverse function to exponentiation
# For eg 1000 = 10 ** 3 ,  logarithm base 10 of 1000 is 3
math.log10(153)
math.floor(math.log10(2345))
```

Logarithm Property: The logarithm of a number tells you what exponent is required to raise the base to in order to get that number. So, log10(2345) tells us what power of 10 equals 2345. <b>
10^3.370 ≈ 2345 <br>
This process involves finding the exponent (approximately 3.370) to which 10 must be raised to equal 2345.

```python
def size_log10base(i):
  return math.floor(math.log10(i)) + 1
size_log10base(153)
```

```python
# Method 3 --> Using string modulo
# "%i" % i: This part formats the integer i as a string using the %i format specifier.
# %i is used to insert integer values into a string. So %i will be replaced by the value of i
def size_mod(i):
  return len("%i" % i) # "%i" is used to convert integer to string using string modulus
size_mod(153)
```

```python
# Method 4 --> Using string.format
def format_size(i):
  return len("{0}".format(i))
format_size(153)
```

### 13. Write a program to find a factorial of a number.

5! (which is read as "5 factorial") equals 5 × 4 × 3 × 2 × 1, and that equals 120.
<br>
So, 5! = 120

```python
num = int(input("Enter the number: "))
factorial  = 1
if num < 0:
  print("Factorial doesn't exist for the Negative numbers")
elif num == 0:
  print("Factorial of 0 is 1")
else:
  for i in range(1, num+1):
    factorial = factorial * i
print(f"Factoral of the Number {num} is : {factorial}")
```

0, 1, 1

0, 1, 1, 2, 3, 5, 8, 13, 21, and so on...
