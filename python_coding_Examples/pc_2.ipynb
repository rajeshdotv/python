### 1.Fibonacci series program in python using iterative method

```python
n = int(input("Enter the nth term: "))
n1 = 0
n2 = 1

count = 0

if n < 0:
  print("Number should be positive")
elif n == 0:
  print(n1)
else:
  while count < n:
    print(n1)
    nth = n1 + n2
    n1 = n2
    n2 = nth
    count += 1
```

```python
# method 2
n = int(input("Please give a number for fibonaaci series: "))
first, second = 0,1
for i in range (0,n):
  if i <=1:
     result = i
  else:
    result = first + second
    first = second
    second = result
  print(result)
```

```python
# method 3 - using recursive method
n = int(input("Please give a number for fibonaaci series: "))
def fibonaaci_recursive(num):
  if num == 0:
    return 0
  elif num == 1:
    return 1
  else:
    return fibonaaci_recursive(num - 1) + fibonaaci_recursive(num - 2)
for i in range(0, n):
  print(fibonaaci_recursive(i))
```

```python
#  x & y  --> In bitwise operations, the & operator performs a bitwise AND operation between two integers
# which means it compares the corresponding bits of two numbers and produces a result where a bit is set (1) if both corresponding bits are set in the operands. Otherwise, it sets the bit to 0.
# x = 1011 #(11 in decimal)
# y = 1101 #(13 in decimal)
# = 1001
# In this example, the result of the bitwise AND operation is 1001 in binary, which is equivalent to decimal 9.
```

```python
# x ^ y --> The bitwise XOR (^) operation is a binary operation that takes two equal-length binary representations and performs the logical XOR operation on each pair of corresponding bits.
# The XOR operation returns a 1 if the bits are different and a 0 if they are the same.
# x = 1010 (decimal: 10)
# y = 1101 (decimal: 13)
#.  = 0111
# When the corresponding bits are different, the result is set to 1.
# When the corresponding bits are the same, the result is set to 0.
# The result of the bitwise XOR operation is 0111 in binary, which is 7 in decimal.
```

```python
x = 10
y = 13
z = x ^ y
z
```

```python
# Method to add 2 numbers without using arithmetic operators (Using Bitwise operators)
def add_numbers(x, y):
  while y!=0:
    temp = x & y # We calculate the carry by performing a bitwise AND operation between x and y. This operation identifies the positions where both x and y have a 1, which indicates a carry.
    x = x ^ y # We add x and y without considering the carry. This is done using the bitwise XOR operation (^). This operation adds the digits without carrying over any values from the previous step.
    y = temp << 1 # We shift the carry one position to the left (<< 1) so that it aligns with the correct position in the addition. This prepares y for the next iteration.
  return x

num1 = 10
num2 = 15
sum_result = add_numbers(num1, num2)
sum_result
```

### 5.Python program to calculate the factorial using iterative approach:

```python
#  If the user input the 5.
#  then the output should be 120 ( 1*2*3*4*5= 120).
n = int(input("Please enter the number: "))
factorial = 1
if n < 0:
  print("Number should be Positive")
elif n == 0:
  print(f"The Factorial of Number {n} is 1")
else:
  for i in range(1, n+1):
    factorial = factorial * i
print(f"The factorial for the number {n} is {factorial}")
```

### 6. Factorial Program in python using recursion with explanation

```python
def factorial_recursive(n):
  if n == 1:
    return n
  else:
    return n*factorial_recursive(n-1)
```

```python
num = int(input("Enter a number to find factorial: "))
if num < 0:
  print("Factorial can't be calculated for negative number. Please enter positive number")
elif num == 0:
  print("Factorial of 0 is 1")
else:
  print(f"Factorial of the numnber {num} is {factorial_recursive(num)}")
```
